```python
import graphviz

# Create a flowchart for split conformal prediction
flowchart = graphviz.Digraph(format="png")

# Define nodes
flowchart.node("A", "Training Data\n(X_train, y_train)", shape="ellipse", style="filled", fillcolor="lightblue")
flowchart.node("B", "Train Model\n(fit on training data)", shape="box", style="filled", fillcolor="lightgray")
flowchart.node("C", "Calibration Data\n(X_calib, y_calib)", shape="ellipse", style="filled", fillcolor="lightgreen")
flowchart.node("D", "Compute Residuals\n|y_calib - f(X_calib)|", shape="box", style="filled", fillcolor="lightgray")
flowchart.node("E", "Estimate Quantile\n(q-hat from residuals)", shape="box", style="filled", fillcolor="lightgray")
flowchart.node("F", "New Test Point\n(X_test)", shape="ellipse", style="filled", fillcolor="lightyellow")
flowchart.node("G", "Compute Prediction Interval\n[f(X_test) ± q-hat]", shape="box", style="filled", fillcolor="lightgray")
flowchart.node("H", "Final Prediction\n(Prediction Interval)", shape="ellipse", style="filled", fillcolor="lightcoral")

# Define edges
flowchart.edge("A", "B")
flowchart.edge("B", "C", label="Split Data")
flowchart.edge("C", "D")
flowchart.edge("D", "E")
flowchart.edge("E", "G", label="q-hat used for interval")
flowchart.edge("F", "G", label="Apply Model")
flowchart.edge("G", "H")

# Render the flowchart
flowchart
```
